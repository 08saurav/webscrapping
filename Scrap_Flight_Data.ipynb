```python
from ratelimit import limits, sleep_and_retry
@sleep_and_retry
@limits(calls=25, period=1)
def myfunc():
    global p
    global q
    global y
    global flag
    p= os.path.getsize('input'+str(y)+'.csv')
    data=pd.read_csv('input'+str(y)+'.csv')
    ct=0
#df=pd.DataFrame(data)
#print(data)
#print(df)
    x=['','','','']
    for i in data.itertuples():
         k=0
         for j in i:
                if(ct%5==0):
                    ct=ct+1
                    continue
                else:
                    #print(j)
                    x[k]=j
                    k=k+1
#             print(ct)
                    ct=ct+1
#             print(x[k-1])
#      print("row= ",ct)
# #src = input("source city name? ")
# #dest = input("destination city name? ")
         temp=x[0].lower()
         src=city.city_code[temp]
         temp=x[1].lower()
         dest=city.city_code[temp]
# #dep= input("date of departure? ")
# #arr= input("date of arrival? ")
# #clas=input("seating class? ")
     #print(type(x[3]))
         url=(get_url(base_url,src,dest,str(x[2]),x[3]))
#      print(url)
         header = {
        'user-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:72.0) Gecko/20100101 Firefox/72.0',
         }
         response = requests.get(url,headers=header)
         if response.status_code != 200:
            raise Exception('API response: {}'.format(response.status_code))
#     return response
#      print(response.status_code)
#          if(flag==1):   
#          print(response)
#          flag=0
         json_data = response.json()
         print(type(json_data))
         flight_data = json_data['data']
         for key, value in flight_data.items():
             print(key)
         len(flight_data['returnflights'])
         airline = []
         deptime = []
         arrtime = []
         stops = []
         flight_code = []
         total_fare = []

         for each in flight_data['onwardflights']:
             print(each)
             airline.append(each['airline'])
             deptime.append(each['deptime'])
             arrtime.append(each['arrtime'])
             stops.append(each['stops'])
             flight_code.append(each['FlHash'])
             total_fare.append(each['fare']['totalfare'])

#      print(airline)
         flight_data = pd.DataFrame({'Airline_Name': airline,
         'Depart_time': deptime,
         'Arri_time': arrtime,
         'Flight_Code': flight_code,
         'Stops': stops,
         'Total_Fare': total_fare
         })
#          print(flight_data.info())
         flight_data['Flight_Code'] = flight_data['Flight_Code'].str.replace('o', '')
         flight_data['Stops'] = flight_data['Stops'].astype(int)
         flight_data.to_csv('./output'+str(y)+'/Flight_'+src+'_to_'+dest+'.csv')
         q=q+os.path.getsize('./output'+str(y)+'/Flight_'+src+'_to_'+dest+'.csv')
         print(q)
         time.sleep(2)
    return 
for i in range(1,11):
    p=0
    q=0
    y=i
    print(y)
    flag=1
    mem = max(memory_usage(proc=myfunc))
    print("Maximum memory used: {} MiB".format(mem))
    print('input file size',p)
    print('output file size',q)
    mem1.append(mem/8)
    inp.append(p/(1024*1024))
    out.append(q/(1024*1024))
```
